### 4. DETEKSI OUTLIER

Tahapan ini bertujuan untuk mendeteksi outlier pada kolom numerik menggunakan metode IQR (Interquartile Range). Untuk setiap fitur seperti umur, pendapatan, skor kredit, dan jumlah pinjaman, dihitung batas bawah dan atas berdasarkan rentang antar kuartil. Data yang berada di luar rentang ini dianggap sebagai outlier. 

Hasil deteksi menunjukkan jumlah dan persentase outlier pada masing-masing kolom, yang penting untuk dipertimbangkan sebelum pelatihan model agar hasil prediksi tidak terdistorsi oleh nilai ekstrem.

```python
# =============================================
# 4. DETEKSI OUTLIER
# =============================================

print("\n" + "="*50)
print("DETEKSI OUTLIER (IQR Method)")
print("="*50)

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

for col in numeric_cols:
    outliers, lower, upper = detect_outliers_iqr(df, col)
    print(f"\n{col.upper()}:")
    print(f"  Batas bawah: {lower:.2f}")
    print(f"  Batas atas: {upper:.2f}")
    print(f"  Jumlah outliers: {len(outliers)} ({len(outliers)/len(df)*100:.1f}%)")
```
